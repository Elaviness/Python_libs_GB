Исходя из чернового графика, результирующими признаками являются:

```python
result_features = ['DistrictId',
 'Rooms',
 'Square',
 'LifeSquare',
 'KitchenSquare',
 'Floor',
 'HouseFloor',
 'HouseYear',
 'Ecology_1',
 'Social_1',
 'Social_2',
 'Social_3',
 'Helthcare_2',
 'Shops_1',
 'LifeSquare_per_Square',
 'Squares_per_Room',
 'Kitchen_per_LifeSquare',
 'Kitchen_per_Square',
 'Shops_1_per_social_1']
len(result_features)
```

## Построение модели

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.ensemble import VotingRegressor

X_result = df[result_features]
y_result = df['Price']

rf_model = RandomForestRegressor(random_state=42, max_depth=30, max_features=7, n_estimators=2250)
rf_model.fit(X_result, y_result)

gb_model = GradientBoostingRegressor(n_estimators=150, max_depth=6, random_state=42)
gb_model.fit(X_result, y_result)

result = VotingRegressor([('rf_model', rf_model), ('gb_model', gb_model)])

result.fit(X_result, y_result)
```

## Прогнозирование на тестовом датасете

```python
X_pred = df_test[result_features]

y_pred = result.predict(X_pred)
preds = pd.DataFrame()
preds['Id'] = X_pred.index
preds['Price'] = y_pred
preds.head()
```

```python
preds.to_csv('MAPugacheva_predictions.csv', index=False)
```
